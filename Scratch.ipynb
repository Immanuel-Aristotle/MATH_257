```python
import numpy as np

def lu_no_pivot(A):
    A = A.astype(float)
    n = A.shape[0]
    L = np.eye(n)
    U = np.zeros_like(A)

    for i in range(n):
        # Upper-triangular part
        for j in range(i, n):
            U[i, j] = A[i, j] - sum(L[i, k] * U[k, j] for k in range(i))
        # Lower-triangular part
        for j in range(i+1, n):
            L[j, i] = (A[j, i] - sum(L[j, k] * U[k, i] for k in range(i))) / U[i, i]

    return L, U

# Example
A = np.array([
  [-2,1,-2],
  [0,-1,0],
  [2,-1,3]
])

L, U = lu_no_pivot(A)

print("A =\n", A)
print("L =\n", L)
print("U =\n", U)
# print("Check A = L @ U:\n", L @ U)
```

```python
import numpy as np
from scipy.linalg import null_space


def project_to_subspace(v, W):
    """
    将向量 v 投影到由 W 的列向量张成的子空间上。
    v: shape (n,)
    W: shape (n, k)
    返回: v 在 span(W) 上的投影向量
    """
    W = np.array(W, dtype=float)
    v = np.array(v, dtype=float)

    # 投影公式: P = W (W^T W)^(-1) W^T v
    P = W @ np.linalg.inv(W.T @ W) @ W.T
    return P


x = np.array([-2, -1, 1])

u1 = np.array([1, -1, -2, 0, -2])
u2 = np.array([0, 0, 1, -2, -1])
U = np.column_stack((u1, u2))  # shape (5, 2)

# Find the orthogonal complement (null space of U^T)
U_perp = null_space(U.T)

import subprocess
import time


def copy_matrix_cells_reverse(matrix, delay=2, use_pbcopy=True):
    """
    依次（倒序）复制矩阵的每个元素到剪贴板。

    参数：
    - matrix: 2D numpy array 或可转换为 array 的对象
    - delay: 每次复制间的时间间隔（秒）
    - use_pbcopy: True 使用 macOS pbcopy；False 只打印内容
    """
    arr = np.array(matrix)
    rows, cols = arr.shape

    # 倒序遍历：从右下角到左上角
    for r in range(0, rows, 1):
      for c in range(0, cols, 1):
    # for r in range(rows - 1, -1, -1):
    #     for c in range(cols - 1, -1, -1):
            val = str(arr[r, c])
            if use_pbcopy:
                subprocess.run("pbcopy", text=True, input=val)
            print(f"Copied cell ({r+1},{c+1}) -> {val}")
            time.sleep(delay)


# print("Orthonormal basis for U⊥:")
# print(U_perp)
# print("Dimension of U⊥ =", U_perp.shape[1])

A = U_perp[:, :2]  # take any two independent vectors
print("Matrix A:\n", A)
print("Rank of A =", np.linalg.matrix_rank(A))
copy_matrix_cells_reverse(A)
```

```python
import numpy as np
def least_sq(A, y):
    # Find the least squares solution to Ax = y
    # by solving Rx = Q^T y for x where A = QR
    Q,R = np.linalg.qr(A)
    x = np.linalg.solve(
        R, Q.T @ y
    )
    return x
def linear_xy(x_y= [(-2, 1), (1, 1), (2, 1), (1, -1), (-1, 0), (-1, -2)]):
  x_line, y_line = [], []
  for x,y in x_y:
    x_line.append(x)
    y_line.append(y)
  A = np.array([
    np.ones(len(x_line)), x_line
  ]).T

  return A.T @ A, A.T @ y_line, least_sq(A, y_line)

x_y = [(2, 4), (-2, 2), (-1, 4), (-1, -3), (2, 4)]
x_line, y_line = [], []
for x,y in x_y:
  x_line.append(x)
  y_line.append(y)
x_line, y_line = np.array(x_line), np.array(y_line)
A = np.array([
  np.ones(x_line.shape),
  x_line,
  x_line**2
]).T
A, A.T @ A, A.T @ y_line, least_sq(A, y_line)
```
