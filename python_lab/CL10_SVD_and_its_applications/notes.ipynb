```python
import numpy as np
import numpy.linalg as la
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])
## important steps
U, S, Vh = la.svd(A)
m, n = U.shape[1], Vh.shape[0]
Sigma = np.zeros((m,n))
np.fill_diagonal(Sigma, S)
U, S, Vh
```

```python
UR, SR, VRt = la.svd(A, full_matrices=False)
UR, SR, VRt
```
