## 3) Colors

In computer graphics, colors are commonly stored by mixing together certain amounts of red, blue, and green (RGB color space). We can thus define RGB colors as vectors in $\mathbb{R}^3$:

$$
\begin{align*}
{\bf red} &= \left(1, 0, 0\right) \\
{\bf green} &= \left(0, 1, 0\right) \\
{\bf blue} &= \left(0, 0, 1\right)
\end{align*}
$$

Which we can then store as numpy arrays:

````python
red   = np.array([1,0,0])
green = np.array([0,1,0])
blue  = np.array([0,0,1])
````

We will use
```python
plt.Circle( (x,y), radius , color=color_array )
```
to create circles with different colors at positions (x,y)

````python
radius = 0.2
````

````python
ax = plt.gca()
ax.set_aspect('equal')
plt.xlim(0,2)
plt.xlabel('x')
plt.ylabel('y')

cred = plt.Circle((0.5,0.5), radius ,color=red)
ax.add_patch(cred)

cgreen = plt.Circle((1,0.5), radius ,color=green)
ax.add_patch(cgreen)

cblue = plt.Circle((1.5,0.5), radius ,color=blue)
ax.add_patch(cblue)
````

We will now write the color ${\bf c}  =  x {\bf u} + y {\bf v} $. For now, we assume that ${\bf u}$ is red and ${\bf v}$ is green. In the next code snippet, we plot circles with position $(x,y)$, radius 0.04, and color defined by  ${\bf c}$. It has a grid of $100$ circles ranging from $[0, 1]$ on the $x$ and $y$ axes.

````python
ax = plt.gca(); ax.set_aspect('equal')
plt.xlim(0,1);plt.ylim(0,1); plt.xlabel('x',fontsize=20); plt.ylabel('y',fontsize=20)

u = red
v = green

points = np.arange(0.,1.1,0.1)

for x in points:
    for y in points:
        # This is the linear combination
        c = x*u + y*v
        circ = plt.Circle((x,y), 0.04, color=c)
        ax.add_patch(circ)
````

Complete the function `make_color_plot` below, so you can use the same code snippet, but with different choices of ${\bf u}$ and ${\bf v}$. The arguments should be the numpy arrays with the colors to be mixed.

Add each color combination (`c`) to the list `colors`.

Once you have the entire list, cast the `colors` list as a numpy array and return it at the end of the function. You can use `list.append()` to add elements to a list. See [here](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists) for more information.

````python
#grade (enter your code in this cell - DO NOT DELETE THIS LINE) 
def make_color_plot(u,v):

    ax = plt.gca(); ax.set_aspect('equal')
    plt.xlim(0,1);plt.ylim(0,1); plt.xlabel('x',fontsize=20); plt.ylabel('y',fontsize=20)

    colors = []
    
    points = np.arange(0.,1.1,0.1)
    # add your code here
    
    for x in points:
        for y in points:
            # This is the linear combination
            c = x*u + y*v
            colors.append(c)
            circ = plt.Circle((x,y), 0.04, color=c)
            ax.add_patch(circ)
        
    return np.array(colors)
````

We can define `purple = (red + blue)/2`. 

If we call `make_color_plot(blue,purple)` below **we get an error**. Why? (Hint: read the last line of the error message.)

````python
purple = (red + blue)/2
make_color_plot(blue,purple);
````

The issue is that when making combinations of `red` and `purple`, we can produce array values **greater than 1**.

The color values should be within the range $\left[0, 1\right]$. We can obtain that using [`np.clip`](https://numpy.org/doc/stable/reference/generated/numpy.clip.html) to force the values into a given range. For example, given the numpy array:

````python
inter = np.array([0.3,0.4,3])
````

We can clip the interval to remain inside pre-defined edges:

````python
np.clip(inter,0,1)
````

For this function, you need to put the computed color for each point in the list `colors`.

Once you have the entire list, cast the `colors` list as a numpy array and return it at the end of the function. You can use `list.append()` to add elements to a list. See [here](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists) for more information.

### Create a music note

Sound data can be represented as a long vector, with each component storing the waveform amplitude at some point in time.

Let's make the sound of the A5 note. (https://en.wikipedia.org/wiki/Piano_key_frequencies)

````python
# We want to have the note played for 0.3 seconds
duration = 0.3
# Define the sampling rate, in hertz (seconds^-1)
DEFAULT_RATE = 44100
# The number of samples needed is
nsamples = int(DEFAULT_RATE*duration)
# The frequency of A5 is 880. 
w = 880
````

We need to create an array of time samples, starting from zero, finishing at `duration`, with `nsamples` points. We will use `np.linspace` for that:

````python
t = np.linspace(0, duration, nsamples)
````

The sound function is:

$f(t) = \sin(w \,2 \pi \,t)$

Define the numpy array `f` corresponding to the points in `t`.

Recall from section 1 that Numpy can perform many operations element-wise for you.

````python
#grade (enter your code in this cell - DO NOT DELETE THIS LINE) 
import math
f = np.sin(w*2*math.pi*t)
````

Check the shape of `f`, does this make sense given the above parameters?


````python
f.shape
````

Plot the sound given by `f`. Inspect a small region of the plot using [`plt.xlim(xlow,xhigh)`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html)

````python
plt.plot(f,'o-')
plt.xlim(0,200)
````

Listen to the sound you created!

````python
ipd.Audio(f,rate=DEFAULT_RATE)
````
