Understanding the progression of the ongoing COVID-19 pandemic has required coordinated efforts of public health organizations at every level around the globe. In this exercise, we will apply principal component analysis (PCA) to compare the coronavirus outbreak in Australia to those in thirteen other countries.



Weekly new case counts per capita for the thirteen other countries are provided as rows of covid_global. These other countries are: Brazil,Canada,China,Ethiopia,Germany,India,Israel,Mexico,Nigeria,Russia,South Africa,United Kingdom,United States (names given in order by names_global). Each row of covid_global represents the time-series of new cases per capita for a given country in weekly increments. The first entry of each row represents the new cases per capita reported over January 22-28, 2020, while the final entry represents those reported over April 13-20, 2022.

First, conduct PCA on the data provided in covid_global. To this end, you will need to compute the average outbreak time-series, shift the data according to this average, then apply singular value decomposition to the shifted data. Store the matrix of right singular vectors from the SVD of the shifted data as V. Remember to compute the reduced SVD by passing the argument full_matrices=False to the function numpy.linalg.svd.

You are also given covid_comp, which represents the time-series data for the outbreak in Australia (and which you will compare to the other thirteen). This has the same format as each row of covid_global but was not included in covid_global so that your PCA was not trained on this testing data. Now, project the time-series data for Australia to the components basis obtained in the previous part. Store this as comp_basis.

Then project the global outbreak data into the components basis, storing the resulting 2d array as global_basis. Finally compare the Australia projection to each of these: compute a 1d array of differences between the Australia projection and that of every other provided country (which you computed as global_basis). Use the Euclidean norm, which is the default of numpy.linalg.norm, for your notion of distance. Store the array of difference values as differences.

Cumulative case count data was pulled from the COVID-19 Data Repository by the Center for Systems Science and Engineering (CSSE) at Johns Hopkins University and is available on Github. Incremental case reportings were computed, then scaled per capita according to estimated country populations from the U.S. Census Bureau, and finally smoothed to represent 2-week rolling averages.

The setup code gives the following variables:

Name	Type	Description
covid_comp	1d numpy array	COVID time-series for the selected comparison country
covid_global	2d numpy array	Array in which each row is a COVID time-series for some country
names_global	list	List of names of countries such that the 
'th name corresponds to the 
'th row of global COVID array
display_pca	function	Plotting function to visualize weights on principal components
Your code snippet should define the following variables:

Name	Type	Description
V	2d numpy array	Matrix of right singular vectors
comp_basis	1d numpy array	Comparison country data projected to component basis
global_basis	2d numpy array	Global countries data projected to component basis
differences	1d numpy array	Array of component-basis distances between the comparison country and every other country

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

# --- 1. Centering the Data ---
# Compute the average time-series (mean across rows/countries, axis=0)
global_avg = np.mean(covid_global, axis=0)

# Shift/Center the global data and the comparison data using the global average
covid_centered = covid_global - global_avg
comp_centered = covid_comp - global_avg

# --- 2. Conduct PCA (SVD) ---
# Apply reduced SVD to the centered global data
U, S, Vt = la.svd(covid_centered, full_matrices=False)

# Store the matrix of right singular vectors (V)
V = Vt.T

# --- 3. Projection into Component Basis ---
# Project the comparison country's centered data onto the PC basis (V)
comp_basis = comp_centered @ V

# Project the global countries' centered data onto the PC basis (V)
global_basis = covid_centered @ V

# --- 4. Compute Differences (Euclidean Distance) ---
# Compute the Euclidean norm (distance) between the comparison country projection
# and every global country projection. The subtraction uses broadcasting.
# axis=1 computes the norm for each row (country).
differences = la.norm(comp_basis - global_basis, axis=1)

# uncomment below to print the global outbreaks by similarity to the comparison country
# ordering is most-similar to least-similar
# print([name for _,name in sorted(zip(differences,names_global),key=lambda p: p[0])])

# uncomment below to visualize your PCA
# display_pca(V, comp_basis, global_basis)
```
